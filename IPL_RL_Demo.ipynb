```python
# Show Rewards CSV Data
import pandas as pd
df = pd.read_csv('training/logs/rewards.csv')
print("Training Log Sample (Last 5 Rows):")
display(df.tail())
```

```python
# Plot Learning Curve
import plotly.express as px

# Group by episode to see overall average reward across all teams
avg_rewards = df.groupby('episode')['TOTAL'].mean().reset_index()

fig = px.line(avg_rewards, x='episode', y='TOTAL', 
              title='Learning Curve: Average Reward per Episode',
              labels={'TOTAL': 'Avg Reward', 'episode': 'Episode'})
fig.update_layout(template='plotly_dark')
fig.show()
```
